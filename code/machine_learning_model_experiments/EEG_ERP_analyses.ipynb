#### Confident Level ERP

```python
import mne
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

# Define file paths for the three confidence conditions
high_confidence_files = sorted(glob('Confident_level_decision_making_ica_remove_eog/High_Confidence/*.fif'))
medium_confidence_files = sorted(glob('Confident_level_decision_making_ica_remove_eog/Medium_Confidence/*.fif'))
low_confidence_files = sorted(glob('Confident_level_decision_making_ica_remove_eog/Low_Confidence/*.fif'))

# Function to load valid (non-empty) epochs
def load_non_empty_epochs(files):
    valid_epochs = []
    for f in files:
        try:
            epochs = mne.read_epochs(f, preload=True)
            if len(epochs) > 0:
                valid_epochs.append(epochs)
            else:
                print(f"No valid epochs in: {f}")
        except Exception as e:
            print(f"Error loading {f}: {e}")
    return valid_epochs

# Load the epochs for each confidence level
epochs_high_confidence = load_non_empty_epochs(high_confidence_files)
epochs_medium_confidence = load_non_empty_epochs(medium_confidence_files)
epochs_low_confidence = load_non_empty_epochs(low_confidence_files)

# Check that we have valid epochs before proceeding
if epochs_high_confidence:
    # Convert all channel names to uppercase
    for ep in epochs_high_confidence + epochs_medium_confidence + epochs_low_confidence:
        ep.rename_channels({name: name.upper() for name in ep.ch_names})

    # Step 1: Focus on Occipital and Parietal Channels
    selected_channels = ['O1', 'O2', 'OZ', 'P1', 'P2', 'PZ', 'PO3', 'PO4', 'PO7', 'PO8', 'P3', 'P4']

    # Step 2: Average the epochs for each subject, focusing on selected channels
    evokeds_high_confidence = [e.copy().pick_channels(selected_channels).crop(tmin=-0.2, tmax=2.0).average() for e in epochs_high_confidence]
    evokeds_medium_confidence = [e.copy().pick_channels(selected_channels).crop(tmin=-0.2, tmax=2.0).average() for e in epochs_medium_confidence]
    evokeds_low_confidence = [e.copy().pick_channels(selected_channels).crop(tmin=-0.2, tmax=2.0).average() for e in epochs_low_confidence]

    # Compute the grand average ERP for each condition
    grand_avg_high_confidence = mne.grand_average(evokeds_high_confidence)
    grand_avg_medium_confidence = mne.grand_average(evokeds_medium_confidence)
    grand_avg_low_confidence = mne.grand_average(evokeds_low_confidence)

    # Step 3: Compute the mean ERP value across selected channels
    mean_high_confidence = grand_avg_high_confidence.data.mean(axis=0)
    mean_medium_confidence = grand_avg_medium_confidence.data.mean(axis=0)
    mean_low_confidence = grand_avg_low_confidence.data.mean(axis=0)

    # Step 4: Visualize the mean ERP values for High, Medium, and Low Confidence in the same figure
    times = grand_avg_high_confidence.times  # Time points for plotting

    plt.figure(figsize=(10, 6))
    plt.plot(times, mean_high_confidence, label='High Confidence', color='blue', linestyle='-', linewidth=1.75)
    plt.plot(times, mean_medium_confidence, label='Medium Confidence', color='orange', linestyle='-', linewidth=1.75)
    plt.plot(times, mean_low_confidence, label='Low Confidence', color='red', linestyle='-', linewidth=1.75)
    plt.axvline(0, color='black', linestyle='--')  # Mark stimulus onset
    plt.title('Mean ERP over Occipital and Parietal Channels (High, Medium, Low Confidence)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()

else:
    print("No valid epochs to process.")
```
